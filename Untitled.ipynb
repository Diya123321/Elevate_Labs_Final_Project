```python
import pandas as pd 
import numpy as np
import os 
import glob
```

```python
file_paths=["Usvideos.csv","CAvideos.csv","DEvideos.csv","FRvideos.csv","GBvideos.csv","INvideos.csv","KRvideos.csv","JPvideos.csv","MXvideos.csv","RUvideos.csv"]
```

```python
dataframes = []
for path in file_paths:
    country = os.path.basename(path)[:2]
    df = pd.read_csv(path, encoding='latin1')
    df["country"] = country
    dataframes.append(df)
```

```python
merged_df = pd.concat(dataframes, ignore_index=True)
```

```python
merged_df.drop_duplicates(inplace=True)
```

```python
merged_df.dropna(inplace=True)
```

```python
merged_df['publish_time'] = pd.to_datetime(merged_df['publish_time'], errors='coerce')
```

```python
merged_df['trending_date'] = pd.to_datetime(merged_df['trending_date'], format='%y.%d.%m', errors='coerce')
```

```python
merged_df.dropna(subset=['publish_time', 'trending_date'], inplace=True)
```

```python
merged_df.to_csv("youtube_cleaned_merged.csv", index=False)
```

```python
import json
```

```python
def load_category_mapping(json_path):
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    
    mapping = {}
    for item in data['items']:
        cat_id = int(item['id'])
        cat_name = item['snippet']['title']
        mapping[cat_id] = cat_name
    return mapping
```

```python
us_mapping = load_category_mapping('US_category_id.json')
in_mapping = load_category_mapping('IN_category_id.json')
mx_mapping = load_category_mapping('MX_category_id.json')
ru_mapping = load_category_mapping('RU_category_id.json')
ca_mapping = load_category_mapping('CA_category_id.json')
de_mapping = load_category_mapping('DE_category_id.json')
fr_mapping = load_category_mapping('FR_category_id.json')
kr_mapping = load_category_mapping('KR_category_id.json')
gb_mapping = load_category_mapping('GB_category_id.json')
jp_mapping = load_category_mapping('JP_category_id.json')


```

```python
df_us = merged_df[merged_df['country'] == 'US'].copy()
df_us['category_name'] = df_us['category_id'].map(us_mapping)
df_in = merged_df[merged_df['country'] == 'IN'].copy()
df_in['category_name'] = df_in['category_id'].map(in_mapping)
df_mx = merged_df[merged_df['country'] == 'MX'].copy()
df_mx['category_name'] = df_mx['category_id'].map(mx_mapping)
df_ru = merged_df[merged_df['country'] == 'RU'].copy()
df_ru['category_name'] = df_ru['category_id'].map(ru_mapping)
```

```python
df_ca = merged_df[merged_df['country'] == 'CA'].copy()
df_ca['category_name'] = df_ca['category_id'].map(ca_mapping)
df_de = merged_df[merged_df['country'] == 'DE'].copy()
df_de['category_name'] = df_de['category_id'].map(de_mapping)
df_fr = merged_df[merged_df['country'] == 'FR'].copy()
df_fr['category_name'] = df_fr['category_id'].map(fr_mapping)
df_kr = merged_df[merged_df['country'] == 'KR'].copy()
df_kr['category_name'] = df_kr['category_id'].map(kr_mapping)
df_gb = merged_df[merged_df['country'] == 'GB'].copy()
df_gb['category_name'] = df_gb['category_id'].map(gb_mapping)
df_jp = merged_df[merged_df['country'] == 'JP'].copy()
df_jp['category_name'] = df_jp['category_id'].map(jp_mapping)
```

```python
final_df = pd.concat([df_us, df_in,df_mx,df_ru,df_ca,df_de,df_fr,df_kr,df_gb,df_jp], ignore_index=True)
```

```python
final_df.to_csv("youtube_cleaned_with_categories.jsonmapped.csv", index=False)
```

```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    try:
        return TextBlob(str(text)).sentiment.polarity
    except:
        return 0
```

```python
final_df['title_sentiment'] = final_df['title'].apply(get_sentiment)
```

```python
final_df['tags_sentiment'] = final_df['tags'].apply(get_sentiment)
```

```python
sentiment_summary = final_df.groupby('category_name')[['title_sentiment', 'tags_sentiment']].mean()
```

```python
final_df = final_df[final_df['views'] > 0]

# Calculate engagement rate
final_df['engagement_rate'] = (
    (final_df['likes'] + final_df['dislikes'] + final_df['comment_count']) 
    / final_df['views']
)


```

```python
engagement_by_category = (
    final_df.groupby('category_name')['engagement_rate'].mean().sort_values(ascending=False)
)

print(engagement_by_category.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
sns.barplot(
    x=engagement_by_category.values,
    y=engagement_by_category.index
)
plt.title("Average Engagement Rate by Category")
plt.xlabel("Engagement Rate")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```
